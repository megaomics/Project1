```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

def extract_spectra(mat_var):
    """Safely extract numerical spectra from MATLAB structures"""
    # If it's already a numeric array
    if isinstance(mat_var, np.ndarray) and mat_var.dtype.kind in 'iuf':
        return mat_var

    # If it's a structured array
    if isinstance(mat_var, np.ndarray) and mat_var.dtype.names:
        for field in mat_var.dtype.names:
            data = mat_var[field][0][0]
            if isinstance(data, np.ndarray) and data.dtype.kind in 'iuf':
                return data

    # If deeply nested
    if isinstance(mat_var, np.ndarray):
        return mat_var[0][0] if mat_var.size == 1 else mat_var

    return mat_var

# --------------------------------------------------
# 1. LOAD AND EXTRACT TRAINING DATA
# --------------------------------------------------
train_data = scipy.io.loadmat('/content/PDAC-plasma-data/trainingspectra.mat')
X_train = np.vstack([
    extract_spectra(train_data['trainingspectra']['normal']),
    extract_spectra(train_data['trainingspectra']['disease']),
    extract_spectra(train_data['trainingspectra']['malignant'])
])
y_train = np.concatenate([np.zeros(58), np.ones(53), np.full(59, 2)])

# --------------------------------------------------
# 2. LOAD AND EXTRACT TEST DATA
# --------------------------------------------------
test_data = scipy.io.loadmat('/content/PDAC-plasma-data/blindedtestspectra.mat')
X_test = extract_spectra(test_data['blindedtestspectra'])

# Ensure correct shape
if X_test.ndim == 1:
    X_test = X_test.reshape(-1, 30142)
elif X_test.ndim > 2:
    X_test = X_test.squeeze()

# --------------------------------------------------
# 3. VERIFICATION AND VISUALIZATION
# --------------------------------------------------
print(f"Training shape: {X_train.shape}")  # Should be (170, 30142)
print(f"Test shape: {X_test.shape}")      # Should be (N, 30142)

# Plot first test sample
plt.figure(figsize=(10,4))
plt.plot(X_test[0] if X_test.shape[0] > 1 else X_test.flatten())
plt.title("First Test Spectrum")
plt.xlabel("Spectral Index")
plt.ylabel("Intensity")
plt.show()
```

```python
import scipy.io
import numpy as np

test_data = scipy.io.loadmat('/content/PDAC-plasma-data/blindedtestspectra.mat')
test_struct = test_data['blindedtestspectra']

# Deep inspection
print("Type:", type(test_struct))
print("Shape:", test_struct.shape)
print("Data type:", test_struct.dtype)
print("First element:", test_struct[0][0])
```

```python
def extract_deep_spectra(mat_var):
    """Recursively extract numerical data from MATLAB structures"""
    if isinstance(mat_var, np.ndarray):
        if mat_var.dtype.names:  # Structured array
            for field in mat_var.dtype.names:
                result = extract_deep_spectra(mat_var[field][0][0])
                if result is not None:
                    return result
        elif mat_var.size == 1:  # Nested single-element array
            return extract_deep_spectra(mat_var[0][0])
        elif mat_var.dtype.kind in 'iuf':  # Numerical data
            return mat_var
    return None

# Extract test spectra
X_test = extract_deep_spectra(test_struct)

# If still not found, try last-resort extraction
if X_test is None:
    X_test = test_struct.ravel()[0]  # Sometimes data is in first element
```

```python
# Ensure proper shape
if X_test is None:
    raise ValueError("No spectra data found in test file")

if X_test.ndim == 1:
    X_test = X_test.reshape(1, -1)  # Single sample

print("Raw extracted shape:", X_test.shape)

# Final check
if X_test.shape[1] != 30142:
    print("Warning: Features dimension mismatch - attempting auto-correction")
    X_test = X_test.reshape(-1, 30142)  # Force correct features
```

```python
import scipy.io
import numpy as np

# 1. Load the data with simplified structure
test_data = scipy.io.loadmat('/content/PDAC-plasma-data/blindedtestspectra.mat',
                            simplify_cells=True)
test_struct = test_data['blindedtestspectra']

# 2. Direct access to the spectra array (based on your structure)
# The spectra is actually stored in the 'classxyz' field
X_test = test_struct['classxyz']

# 3. Convert and reshape
X_test = np.asarray(X_test, dtype=np.float64)

# Reshape to (samples × features)
if X_test.ndim == 1:
    X_test = X_test.reshape(1, -1)  # Single sample
elif X_test.shape[1] != 30142:
    X_test = X_test.T  # Transpose if needed

print("Final test data shape:", X_test.shape)
assert X_test.shape[1] == 30142, "Feature dimension mismatch"

# 4. Verification
print("Data range:", X_test.min(), X_test.max())
plt.figure(figsize=(10,4))
plt.plot(X_test[0])
plt.title("First Test Spectrum")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential([
    Dense(64, activation='relu', input_shape=(30142,)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')  # 3 classes: normal, benign, malignant
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train_scaled, y_train,
                    epochs=100,
                    batch_size=32,
                    validation_split=0.2,
                    verbose=1)
```

```python
# Predict probabilities
y_probs = model.predict(X_test_scaled)

# Get class predictions (0=normal, 1=benign, 2=malignant)
y_pred = np.argmax(y_probs, axis=1)

# If you had test labels (y_test), you could use:
# from sklearn.metrics import classification_report
# print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.legend()

plt.subplot(122)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.legend()
plt.show()
```

```python
# Get weights from first layer
weights = model.layers[0].get_weights()[0]  # Shape: (30142, 64)

# Calculate feature importance
importance = np.mean(np.abs(weights), axis=1)

# Plot top important regions
top_indices = np.argsort(importance)[-20:]
plt.figure(figsize=(12,6))
plt.bar(range(20), importance[top_indices])
plt.xticks(range(20), top_indices, rotation=45)
plt.xlabel('Spectral Index')
plt.ylabel('Importance Score')
plt.title('Top 20 Important Spectral Regions')
plt.show()
```

```python
def predict_cancer(new_spectrum):
    """API-ready prediction function"""
    spectrum_scaled = scaler.transform(new_spectrum.reshape(1, -1))
    proba = model.predict(spectrum_scaled)[0]
    return {'normal': float(proba[0]),
            'benign': float(proba[1]),
            'malignant': float(proba[2])}
```
